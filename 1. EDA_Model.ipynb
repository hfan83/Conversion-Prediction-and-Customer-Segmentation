Một vài quan sát với tập dữ liệu này:
- Khoảng Min Max của các features không có sự bất thường 
- Phân vị thứ 2/Median (50%) là điểm giữa của dữ liệu gần khớp với mean cho thấy dữ liệu gần như đối xứng
- Phân vị thứ 3 (75%) và khoảng Max tương đối gần nhau cho thấy dữ liệu không có giá trị ngoại lệ


```python
def plot_distribution(data, numeric_columns, n_cols=2):
    """
    Vẽ biểu đồ histogram và các thống kê cơ bản cho các biến số
    
    Parameters:
    data: DataFrame chứa dữ liệu
    numeric_columns: List các cột cần phân tích
    n_cols: Số cột trong subplot grid
    """
    # Tính số hàng cần thiết cho subplot grid
    n_rows = int(np.ceil(len(numeric_columns) / n_cols))
    
    # Tạo figure với kích thước phù hợp
    fig = plt.figure(figsize=(15, 5*n_rows))

    
    for idx, column in enumerate(numeric_columns, 1):
        # Tạo subplot
        ax = plt.subplot(n_rows, n_cols, idx)
        
        # Vẽ histogram với kernel density estimation
        sns.histplot(data=data, x=column, kde=True, ax=ax, )
        
        # Tính các thống kê cơ bản
        mean_val = data[column].mean()
        median_val = data[column].median()
        skew_val = data[column].skew()
        kurt_val = data[column].kurtosis()
        
        # Vẽ đường vertical line cho mean và median
        ax.axvline(mean_val, color='red', linestyle='--', alpha=0.8, label=f'Mean: {mean_val:.2f}')
        ax.axvline(median_val, color='green', linestyle='--', alpha=0.8, label=f'Median: {median_val:.2f}')
        
        # Thêm tiêu đề và thống kê
        ax.set_title(f'Phân phối của {column}\n' + 
                    f'Skewness: {skew_val:.2f}, Kurtosis: {kurt_val:.2f}', 
                    pad=20)
        
        # Thêm legend
        ax.legend()
        
        # Thêm grid
        ax.grid(True, alpha=0.3)
        
        # Tối ưu hóa số lượng bins dựa trên dữ liệu
        if data[column].nunique() > 50:
            # Sử dụng Freedman-Diaconis rule để tính số bins tối ưu
            q75, q25 = np.percentile(data[column], [75, 25])
            iqr = q75 - q25
            bin_width = 2 * iqr / (len(data[column]) ** (1/3))
            if bin_width > 0:
                bins = int((data[column].max() - data[column].min()) / bin_width)
                ax.clear()
                sns.histplot(data=data, x=column, bins=min(bins, 50), kde=True, ax=ax)
                ax.axvline(mean_val, color='red', linestyle='--', alpha=0.8, label=f'Mean: {mean_val:.2f}')
                ax.axvline(median_val, color='green', linestyle='--', alpha=0.8, label=f'Median: {median_val:.2f}')
                ax.legend()
                ax.grid(True, alpha=0.3)
    
    # Thêm tiêu đề chung
    fig.suptitle('Phân phối của các biến số', fontsize=16, y=1.02)
    
    # Điều chỉnh layout
    plt.tight_layout()
    
    return plt

# Danh sách các cột cần phân tích
numeric_columns = [
    'Age', 'Income', 'AdSpend', 'ClickThroughRate', 
    'ConversionRate', 'WebsiteVisits', 'PagesPerVisit', 'TimeOnSite',  'SocialShares', 'EmailOpens', 'PreviousPurchases', 'LoyaltyPoints'
]


```

- Giá trị mean và median ở các phân phối gần như bằng nhau cho thấy phân phối đối xứng 
- Không có hiện tượng Skewness (độ lệch) rõ rệt trong bất kỳ biến nào

# 2. Phân tích các biến phân loại với biến mục tiêu Conversion

```python
def create_conversion_stacked_chart(data):
    # Tính toán tỷ lệ chuyển đổi cho từng nhóm
    conversion_counts = data['Conversion'].value_counts()
    conversion_percentages = (conversion_counts / len(data) * 100).round(2)
    
    # Tạo DataFrame cho biểu đồ
    df_plot = pd.DataFrame({
        'Status': ['Không chuyển đổi', 'Chuyển đổi'],
        'Percentage': conversion_percentages,
    })
    
    # Tạo biểu đồ
    plt.figure(figsize=(10, 6))
    
    # Vẽ biểu đồ chồng
    plt.bar(0, df_plot['Percentage'][0], color='#ff9999', label='Không chuyển đổi')
    plt.bar(0, df_plot['Percentage'][1], bottom=df_plot['Percentage'][0], color='#66b3ff', label='Chuyển đổi')
    
    # Chỉnh sửa định dạng
    plt.title('Tỷ lệ chuyển đổi của khách hàng', pad=20, fontsize=14)
    plt.ylabel('Phần trăm (%)', fontsize=14)
    plt.xticks([])  # Ẩn trục x vì chỉ có một cột
    
    # Thêm nhãn phần trăm vào các phần của biểu đồ
    for i, percentage in enumerate(df_plot['Percentage']):
        if i == 0:
            y = percentage / 2
        else:
            y = df_plot['Percentage'][0] + percentage / 2
        plt.text(0, y, f'{percentage:.1f}%', ha='center', va='center')
    
    plt.legend(loc='upper right')
    
    # Thêm lưới để dễ đọc
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    
    # Điều chỉnh khoảng cách các cạnh
    plt.margins(x=0.4)
    
    return plt
```

```python
plot = create_conversion_stacked_chart(df)
plot.show()
```

```python
def plot_conversion_by_category(data, category_column):
    """
    Tạo biểu đồ chồng thể hiện tỷ lệ chuyển đổi theo biến phân loại
    
    Parameters:
    data: DataFrame chứa dữ liệu
    category_column: Tên cột chứa biến phân loại cần phân tích
    """
    # Tính toán số lượng và tỷ lệ cho mỗi category
    conversion_by_category = pd.crosstab(
        data[category_column], 
        data['Conversion'],
        normalize='index'
    ) * 100
    
    # Đặt tên cho các cột
    conversion_by_category.columns = ['Không chuyển đổi', 'Chuyển đổi']
    
    # Thêm cột count để hiển thị số lượng mẫu
    category_counts = data[category_column].value_counts()
    conversion_by_category['Total'] = category_counts
    
    # Sắp xếp theo tỷ lệ chuyển đổi giảm dần
    conversion_by_category = conversion_by_category.sort_values('Chuyển đổi', ascending=True)
    
    # Tạo biểu đồ
    plt.figure(figsize=(12, 6))
    
    # Vẽ biểu đồ chồng
    bars = plt.barh(range(len(conversion_by_category)), 
                   conversion_by_category['Không chuyển đổi'],
                   color='#ff9999',
                   label='Không chuyển đổi')
    
    bars2 = plt.barh(range(len(conversion_by_category)), 
                    conversion_by_category['Chuyển đổi'],
                    left=conversion_by_category['Không chuyển đổi'],
                    color='#66b3ff',
                    label='Chuyển đổi')
    
    # Thêm nhãn phần trăm
    for i in range(len(conversion_by_category)):
        # Nhãn cho không chuyển đổi
        plt.text(conversion_by_category['Không chuyển đổi'][i]/2,
                i,
                f'{conversion_by_category["Không chuyển đổi"][i]:.1f}%',
                ha='center',
                va='center')
        
        # Nhãn cho chuyển đổi
        plt.text(conversion_by_category['Không chuyển đổi'][i] + 
                conversion_by_category['Chuyển đổi'][i]/2,
                i,
                f'{conversion_by_category["Chuyển đổi"][i]:.1f}%',
                ha='center',
                va='center')
    
    # Chỉnh sửa định dạng
    plt.title(f'Tỷ lệ chuyển đổi theo {category_column}', pad=20, fontsize=14)
    plt.xlabel('Phần trăm (%)', fontsize=12)
    
    # Đặt nhãn trục y với số lượng mẫu
    plt.yticks(range(len(conversion_by_category)),
              [f'{idx} (n={conversion_by_category["Total"][idx]})' 
               for idx in conversion_by_category.index])
    
    plt.legend(loc='lower right')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    
    # Điều chỉnh khoảng cách
    plt.tight_layout()
    
    return plt

# Sử dụng hàm để vẽ biểu đồ cho các biến phân loại khác nhau
def plot_all_categorical_conversions(data):
    """
    Vẽ biểu đồ cho tất cả các biến phân loại quan trọng
    """
    categorical_columns = ['Gender', 'CampaignType', 'CampaignChannel']
    
    for column in categorical_columns:
        plt.figure()
        plot_conversion_by_category(data, column)
        plt.show()
```

```python
def create_age_income_bins(data):
    """
    Tạo các nhóm tuổi và thu nhập từ dữ liệu theo phân khúc mới
    """
    # Tạo copy của DataFrame
    df = data.copy()
    
    # Tạo bins cho tuổi theo 4 nhóm chỉ định
    age_bins = [18, 30, 40, 55, 70]
    age_labels = ['18-29', '30-39', '40-54', '55-70']
    df['AgeGroup'] = pd.cut(df['Age'], 
                           bins=age_bins, 
                           labels=age_labels, 
                           right=False)
    
    # Tạo bins cho thu nhập theo 4 nhóm chỉ định
    income_bins = [20000, 40000, 70000, 100000, float('inf')]
    income_labels = ['20k-40k', '40k-70k', '70k-100k', '100k+']
    df['IncomeGroup'] = pd.cut(df['Income'], 
                              bins=income_bins, 
                              labels=income_labels, 
                              right=False)
    
    return df

def plot_conversion_by_group(data, group_column, title):
    """
    Vẽ biểu đồ cột chồng cho tỷ lệ chuyển đổi theo nhóm
    """
    # Tính tỷ lệ chuyển đổi cho mỗi nhóm
    conversion_by_group = pd.crosstab(
        data[group_column], 
        data['Conversion'],
        normalize='index'
    ) * 100
    
    # Đặt tên cho các cột
    conversion_by_group.columns = ['Không chuyển đổi', 'Chuyển đổi']
    
    # Thêm cột count để hiển thị số lượng mẫu
    group_counts = data[group_column].value_counts()
    conversion_by_group['Total'] = group_counts
    
    # Tạo biểu đồ
    plt.figure(figsize=(12, 6))
    
    # Vẽ biểu đồ cột chồng
    bars = plt.bar(range(len(conversion_by_group)), 
                  conversion_by_group['Không chuyển đổi'],
                  color='#ff9999',
                  label='Không chuyển đổi')
    
    bars2 = plt.bar(range(len(conversion_by_group)), 
                   conversion_by_group['Chuyển đổi'],
                   bottom=conversion_by_group['Không chuyển đổi'],
                   color='#66b3ff',
                   label='Chuyển đổi')
    
    # Thêm nhãn phần trăm và số lượng mẫu
    for i in range(len(conversion_by_group)):
        # Nhãn cho không chuyển đổi
        plt.text(i, 
                conversion_by_group['Không chuyển đổi'][i]/2,
                f'{conversion_by_group["Không chuyển đổi"][i]:.1f}%\n(n={conversion_by_group["Total"][i]})',
                ha='center',
                va='center')
        
        # Nhãn cho chuyển đổi
        plt.text(i,
                conversion_by_group['Không chuyển đổi'][i] + 
                conversion_by_group['Chuyển đổi'][i]/2,
                f'{conversion_by_group["Chuyển đổi"][i]:.1f}%',
                ha='center',
                va='center')
    
    # Chỉnh sửa định dạng
    plt.title(title, pad=20, fontsize=14)
    plt.ylabel('Phần trăm (%)', fontsize=12)
    plt.xticks(range(len(conversion_by_group)), 
              conversion_by_group.index, 
              rotation=45)
    
    plt.legend(loc='upper right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Điều chỉnh khoảng cách
    plt.tight_layout()
    
    return plt

def analyze_conversion_rates(data):
    """
    Phân tích và vẽ biểu đồ tỷ lệ chuyển đổi cho cả nhóm tuổi và thu nhập
    """
    # Tạo các nhóm
    df_binned = create_age_income_bins(data)
    
    # Vẽ biểu đồ cho nhóm tuổi
    plt.figure(1)
    plot_conversion_by_group(df_binned, 'AgeGroup', 
                           'Tỷ lệ chuyển đổi theo nhóm tuổi')
    
    # Vẽ biểu đồ cho nhóm thu nhập
    plt.figure(2)
    plot_conversion_by_group(df_binned, 'IncomeGroup', 
                           'Tỷ lệ chuyển đổi theo nhóm thu nhập')
    
    # Tạo bảng thống kê chi tiết
    stats = []
    for group in ['AgeGroup', 'IncomeGroup']:
        conversion_stats = pd.crosstab(df_binned[group], 
                                     df_binned['Conversion'],
                                     normalize='index') * 100
        conversion_stats['Total_Samples'] = df_binned[group].value_counts()
        stats.append(conversion_stats)
    
    return stats
```

```python
stats = analyze_conversion_rates(df)
```

```python

def create_purchase_loyalty_segments(data):
    """
    Tạo các phân đoạn cho số lần mua trước đây và điểm tích lũy
    """
    df = data.copy()
    
    # Phân đoạn PreviousPurchases
    purchase_bins = [0, 1, 5, 9]
    purchase_labels = ['0-1', '2-5', '6-9']
    df['PurchaseGroup'] = pd.cut(df['PreviousPurchases'], 
                                bins=purchase_bins,
                                labels=purchase_labels,
                                right=True)
    
    # Phân đoạn LoyaltyPoints
    loyalty_bins = [0, 1000, 3000, 5000]  # Chia thành ba khoảng: 0-1000, 1000-3000, 3000-5000
    loyalty_labels = ['0-1k', '1k-3k', '3k-5k']  # Nhãn tương ứng với các khoảng
    df['LoyaltyGroup'] = pd.cut(df['LoyaltyPoints'],
                            bins=loyalty_bins,
                            labels=loyalty_labels,
                            right=True)

    
    return df

def plot_conversion_by_segment(data, segment_column, title):
    """
    Vẽ biểu đồ cột chồng cho tỷ lệ chuyển đổi theo phân đoạn
    """
    # Tính tỷ lệ chuyển đổi cho mỗi phân đoạn
    conversion_by_segment = pd.crosstab(
        data[segment_column],
        data['Conversion'],
        normalize='index'
    ) * 100
    
    # Đặt tên cho các cột
    conversion_by_segment.columns = ['Không chuyển đổi', 'Chuyển đổi']
    
    # Thêm cột count để hiển thị số lượng mẫu
    segment_counts = data[segment_column].value_counts()
    conversion_by_segment['Total'] = segment_counts
    
    # Sắp xếp theo index để đảm bảo thứ tự đúng
    conversion_by_segment = conversion_by_segment.sort_index()
    
    # Tạo biểu đồ
    plt.figure(figsize=(12, 6))
    
    # Vẽ biểu đồ cột chồng
    bars = plt.bar(range(len(conversion_by_segment)),
                  conversion_by_segment['Không chuyển đổi'],
                  color='#ff9999',
                  label='Không chuyển đổi')
    
    bars2 = plt.bar(range(len(conversion_by_segment)),
                   conversion_by_segment['Chuyển đổi'],
                   bottom=conversion_by_segment['Không chuyển đổi'],
                   color='#66b3ff',
                   label='Chuyển đổi')
    
    # Thêm nhãn phần trăm và số lượng mẫu
    for i in range(len(conversion_by_segment)):
        # Nhãn cho không chuyển đổi
        plt.text(i,
                conversion_by_segment['Không chuyển đổi'][i]/2,
                f'{conversion_by_segment["Không chuyển đổi"][i]:.1f}%\n(n={conversion_by_segment["Total"][i]})',
                ha='center',
                va='center')
        
        # Nhãn cho chuyển đổi
        plt.text(i,
                conversion_by_segment['Không chuyển đổi'][i] +
                conversion_by_segment['Chuyển đổi'][i]/2,
                f'{conversion_by_segment["Chuyển đổi"][i]:.1f}%',
                ha='center',
                va='center')
    
    # Chỉnh sửa định dạng
    plt.title(title, pad=20, fontsize=14)
    plt.ylabel('Phần trăm (%)', fontsize=12)
    plt.xlabel('Phân đoạn', fontsize=12)
    plt.xticks(range(len(conversion_by_segment)),
               conversion_by_segment.index,
               rotation=0)
    
    plt.legend(loc='upper right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Điều chỉnh khoảng cách
    plt.tight_layout()
    
    return plt, conversion_by_segment

def analyze_segments_and_correlations(data):
    """
    Phân tích chi tiết các phân đoạn và mối tương quan với tỷ lệ chuyển đổi
    """
    # Tạo các phân đoạn
    df_segmented = create_purchase_loyalty_segments(data)
    
    # Vẽ biểu đồ và tính thống kê cho số lần mua trước đây
    plt.figure(1)
    plt_purchase, stats_purchase = plot_conversion_by_segment(
        df_segmented,
        'PurchaseGroup',
        'Tỷ lệ chuyển đổi theo số lần mua trước đây'
    )
    
    # Vẽ biểu đồ và tính thống kê cho điểm tích lũy
    plt.figure(2)
    plt_loyalty, stats_loyalty = plot_conversion_by_segment(
        df_segmented,
        'LoyaltyGroup',
        'Tỷ lệ chuyển đổi theo điểm tích lũy'
    )
    
    # Tính correlation giữa các biến gốc và Conversion
    correlations = pd.DataFrame({
        'PreviousPurchases': [data['PreviousPurchases'].corr(data['Conversion'])],
        'LoyaltyPoints': [data['LoyaltyPoints'].corr(data['Conversion'])]
    }, index=['Correlation with Conversion'])
    
    return {
        'purchase_stats': stats_purchase,
        'loyalty_stats': stats_loyalty,
        'correlations': correlations
    }
```

```python
results = analyze_segments_and_correlations(df)
```

Nhận xét:

    + Tương tự với phân đoạn theo điểm tích lũy, khách hàng từ 0 - 1k điểm tích lũy có tỉ lệ chuyển đổi thấp nhất trong 3 nhóm khách hàng.

```python
# Bước 1: Phân loại 'AdSpend' thành các nhóm giá trị từ 100 đến 10000
bins_adspend = [100, 1000, 3000, 5000, 7000, 10000]
labels_adspend = ['100-1000', '1000-3000', '3000-5000', '5000-7000', '7000-10000']
df['AdSpendGroup'] = pd.cut(df['AdSpend'], bins=bins_adspend, labels=labels_adspend, right=False)

# Bước 2: Tính toán số lượng khách hàng chuyển đổi và không chuyển đổi theo nhóm 'AdSpend'
adspend_conversion = df.groupby(['AdSpendGroup', 'Conversion']).size().unstack(fill_value=0)

# Bước 3: Tính toán phần trăm khách hàng chuyển đổi và không chuyển đổi trong mỗi nhóm
adspend_conversion_percent = adspend_conversion.div(adspend_conversion.sum(axis=1), axis=0) * 100

# Bước 4: Vẽ biểu đồ chồng (stacked bar chart) với phần trăm
ax = adspend_conversion_percent.plot(kind='bar', stacked=True, color=['#ff9999', '#66b3ff'], figsize=(10,6))

# Thêm nhãn phần trăm vào từng cột
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.1f}%', (x + width/2, y + height/2), ha='center', va='center')

# Thêm tiêu đề và nhãn
plt.title('Phân phối AdSpend theo tỷ lệ chuyển đổi (%)')
plt.xlabel('AdSpend (Nhóm)')
plt.ylabel('Phần trăm (%)')
plt.xticks(rotation=45)

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

```python
def create_age_income_loyalty_groups(data):
    """
    Tạo nhóm tuổi, thu nhập và loyalty points
    """
    df = data.copy()
   
    # Tạo nhóm tuổi
    age_bins = [18, 30, 40, 55, 70]  # Định nghĩa các ranh giới nhóm tuổi
    age_labels = ['18-29', '30-39', '40-54', '55-70']  # Tên các nhóm tuổi
    df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

    # Tạo nhóm thu nhập
    income_bins = [20000, 40000, 70000, 100000, float('inf')]  # Ranh giới các nhóm thu nhập
    income_labels = ['20k-40k', '40k-70k', '70k-100k', '100k+']  # Tên các nhóm thu nhập
    df['IncomeGroup'] = pd.cut(df['Income'], bins=income_bins, labels=income_labels, right=False)
    
    # Tạo nhóm loyalty points (chia thành 3 nhóm)
    loyalty_bins = [0, 1000, 3000, 5000]  # Ranh giới các nhóm loyalty points
    loyalty_labels = ['0-1000', '1000-3000', '3000-5000']  # Tên các nhóm loyalty points
    df['LoyaltyGroup'] = pd.cut(df['LoyaltyPoints'], bins=loyalty_bins, labels=loyalty_labels, right=False)
    
    # Tạo nhóm previous purchase
    purchase_bins = [0, 1, 5, 9]  # Ranh giới các nhóm previous purchase
    purchase_labels = ['0-1', '2-5', '6-9']  # Tên các nhóm previous purchase
    df['PurchaseGroup'] = pd.cut(df['PreviousPurchases'], bins=purchase_bins, labels=purchase_labels, right=False)
    
    return df
```

```python
def plot_conversion_distribution_by_age_and_income(data):
    """
    Vẽ ba biểu đồ phân phối số khách hàng theo nhóm tuổi, nhóm thu nhập và nhóm điểm trung thành,
    đồng thời hiển thị phần trăm chuyển đổi 0 và 1 cho từng loại chiến dịch tiếp thị.
    """
    # Tạo nhóm tuổi, thu nhập và loyalty points
    df = create_age_income_loyalty_groups(data)

    # Danh sách các loại chiến dịch tiếp thị
    campaign_types = df['CampaignType'].unique()
    
    # Hàm tính và thêm phần trăm vào cột
    def add_percentage_labels(ax, conversion_counts):
        for container in ax.containers:
            labels = []
            for i, rect in enumerate(container):
                height = rect.get_height()
                total = conversion_counts.iloc[i % len(conversion_counts)].sum()
                if total > 0:
                    percent = height / total * 100
                    labels.append(f'{percent:.1f}%')
                else:
                    labels.append('0%')
            ax.bar_label(container, labels=labels, label_type='center')

    # Vẽ biểu đồ theo nhóm tuổi
    fig, axes = plt.subplots(2, 2, figsize=(15, 10), sharey=True)
    axes = axes.flatten()

    for i, campaign in enumerate(campaign_types):
        campaign_data = df[df['CampaignType'] == campaign]
        
        # Đếm số lượng khách hàng theo nhóm tuổi và tỷ lệ chuyển đổi
        conversion_counts_age = campaign_data.groupby(['AgeGroup', 'Conversion']).size().unstack(fill_value=0)

        # Vẽ stacked bar cho tỷ lệ chuyển đổi theo nhóm tuổi
        ax = conversion_counts_age.plot(kind='bar', stacked=True, ax=axes[i], color=['#FF9999', '#66B3FF'])
        
        # Thêm nhãn phần trăm cho các cột
        add_percentage_labels(ax, conversion_counts_age)
        
        # Định dạng biểu đồ
        ax.set_title(f'Chuyển đổi theo Nhóm Tuổi cho {campaign}', fontsize=14)
        ax.set_xlabel('Nhóm Tuổi', fontsize=12)
        ax.set_ylabel('Số lượng khách hàng', fontsize=12)
        ax.set_xticklabels(conversion_counts_age.index, rotation=45)
        ax.legend(title='Chuyển đổi', labels=['Không chuyển đổi (0)', 'Chuyển đổi (1)'], bbox_to_anchor=(1.05, 1), loc='upper left')
        ax.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()

    # Vẽ biểu đồ theo nhóm thu nhập
    fig, axes = plt.subplots(2, 2, figsize=(15, 10), sharey=True)
    axes = axes.flatten()

    for i, campaign in enumerate(campaign_types):
        campaign_data = df[df['CampaignType'] == campaign]
        
        # Đếm số lượng khách hàng theo nhóm thu nhập và tỷ lệ chuyển đổi
        conversion_counts_income = campaign_data.groupby(['IncomeGroup', 'Conversion']).size().unstack(fill_value=0)

        # Vẽ stacked bar cho tỷ lệ chuyển đổi theo nhóm thu nhập
        ax = conversion_counts_income.plot(kind='bar', stacked=True, ax=axes[i], color=['#FF9999', '#66B3FF'])
        
        # Thêm nhãn phần trăm cho các cột
        add_percentage_labels(ax, conversion_counts_income)
        
        # Định dạng biểu đồ
        ax.set_title(f'Chuyển đổi theo Nhóm Thu Nhập cho {campaign}', fontsize=14)
        ax.set_xlabel('Nhóm Thu Nhập', fontsize=12)
        ax.set_ylabel('Số lượng khách hàng', fontsize=12)
        ax.set_xticklabels(conversion_counts_income.index, rotation=45)
        ax.legend(title='Chuyển đổi', labels=['Không chuyển đổi (0)', 'Chuyển đổi (1)'], bbox_to_anchor=(1.05, 1), loc='upper left')
        ax.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()

    # Vẽ biểu đồ theo nhóm điểm trung thành
    fig, axes = plt.subplots(2, 2, figsize=(15, 10), sharey=True)
    axes = axes.flatten()

    for i, campaign in enumerate(campaign_types):
        campaign_data = df[df['CampaignType'] == campaign]
        
        # Đếm số lượng khách hàng theo nhóm điểm trung thành và tỷ lệ chuyển đổi
        conversion_counts_loyalty = campaign_data.groupby(['LoyaltyGroup', 'Conversion']).size().unstack(fill_value=0)

        # Vẽ stacked bar cho tỷ lệ chuyển đổi theo nhóm điểm trung thành
        ax = conversion_counts_loyalty.plot(kind='bar', stacked=True, ax=axes[i], color=['#FF9999', '#66B3FF'])
        
        # Thêm nhãn phần trăm cho các cột
        add_percentage_labels(ax, conversion_counts_loyalty)
        
        # Định dạng biểu đồ
        ax.set_title(f'Chuyển đổi theo Nhóm Điểm Trung Thành cho {campaign}', fontsize=14)
        ax.set_xlabel('Nhóm Điểm Trung Thành', fontsize=12)
        ax.set_ylabel('Số lượng khách hàng', fontsize=12)
        ax.set_xticklabels(conversion_counts_loyalty.index, rotation=45)
        ax.legend(title='Chuyển đổi', labels=['Không chuyển đổi (0)', 'Chuyển đổi (1)'], bbox_to_anchor=(1.05, 1), loc='upper left')
        ax.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()

    # Vẽ biểu đồ theo tần suất mua trước đó
    fig, axes = plt.subplots(2, 2, figsize=(15, 10), sharey=True)
    axes = axes.flatten()

    for i, campaign in enumerate(campaign_types):
        campaign_data = df[df['CampaignType'] == campaign]
        
        # Đếm số lượng khách hàng theo tần suất mua trước đó và tỷ lệ chuyển đổi
        conversion_counts_loyalty = campaign_data.groupby(['PurchaseGroup', 'Conversion']).size().unstack(fill_value=0)

        # Vẽ stacked bar cho tỷ lệ chuyển đổi theo tần suất mua trước đó
        ax = conversion_counts_loyalty.plot(kind='bar', stacked=True, ax=axes[i], color=['#FF9999', '#66B3FF'])
        
        # Thêm nhãn phần trăm cho các cột
        add_percentage_labels(ax, conversion_counts_loyalty)
        
        # Định dạng biểu đồ
        ax.set_title(f'Chuyển đổi theo theo Tần suất mua trước đó cho {campaign}', fontsize=14)
        ax.set_xlabel('Tần suất mua trước đó', fontsize=12)
        ax.set_ylabel('Số lượng khách hàng', fontsize=12)
        ax.set_xticklabels(conversion_counts_loyalty.index, rotation=45)
        ax.legend(title='Chuyển đổi', labels=['Không chuyển đổi (0)', 'Chuyển đổi (1)'], bbox_to_anchor=(1.05, 1), loc='upper left')
        ax.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()
```

# 3. Phân tích hành vi tương tác của khách hàng theo các biến phân loại

```python
def plot_metrics_by_group(data, group_column, metrics, title):
    """
    Vẽ biểu đồ cột cho các chỉ số theo nhóm
    """
    # Tính giá trị trung bình cho mỗi chỉ số theo nhóm
    means = data.groupby(group_column)[metrics].mean()
    
    # Tính số lượng mẫu trong mỗi nhóm
    counts = data[group_column].value_counts()
    
    # Số lượng chỉ số và nhóm
    n_metrics = len(metrics)
    n_groups = len(means.index)
    
    # Tạo figure với kích thước phù hợp
    plt.figure(figsize=(15, 8))
    
    # Tạo vị trí cho các cột
    width = 0.8 / n_metrics
    x = np.arange(n_groups)

# Danh sách màu sắc cho các cột
    custom_colors = [   '#779ECB', '#62B6CB', '#1B4965', '#CAE9FF', '#5FA8D3',
    '#0C2D48', '#2E5984', '#8FC1E3']


    
    # Vẽ biểu đồ cột cho từng chỉ số
    for i, metric in enumerate(metrics):
        position = x + (i - n_metrics/2 + 0.5) * width
        bars = plt.bar(position, means[metric], width,
                      label=metric,
                      color = custom_colors[i %len(custom_colors)])
        
        # Thêm giá trị lên đầu cột
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.2f}',
                    ha='center', va='bottom', rotation=90)
    
    # Chỉnh sửa định dạng
    plt.title(f'{title}\n(Số mẫu trong mỗi nhóm được hiển thị trong ngoặc)', 
              pad=20, fontsize=14)
    plt.xlabel(group_column, fontsize=12)
    plt.ylabel('Giá trị trung bình', fontsize=12)
    
    # Thêm số lượng mẫu vào nhãn trục x
    plt.xticks(x, [f'{idx}\n(n={counts[idx]})' for idx in means.index], 
               rotation=45, ha='right')
    
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Điều chỉnh khoảng cách
    plt.tight_layout()
    
    return plt

def analyze_metrics_by_groups(data):
    """
    Phân tích và vẽ biểu đồ cho tất cả các nhóm
    """
    # Tạo nhóm tuổi, thu nhập và loyalty points
    df = create_age_income_loyalty_groups(data)
    
    # Danh sách các chỉ số cần phân tích
    metrics = ['TimeOnSite', 'PagesPerVisit', 'WebsiteVisits',
              'SocialShares', 'EmailOpens', 'EmailClicks', 'ClickThroughRate']
    
    # Danh sách các nhóm cần phân tích
    groups = {
        'Gender': 'Giới tính',
        'AgeGroup': 'Nhóm tuổi',
        'IncomeGroup': 'Nhóm thu nhập',
        'LoyaltyGroup': 'Nhóm loyalty points',
        'PurchaseGroup': 'Tần suất mua trước đó',
        'CampaignType': 'Loại chiến dịch',
        'CampaignChannel': 'Kênh marketing',
        
    }
    
    # Tạo bảng thống kê tổng hợp
    summary_stats = pd.DataFrame()
    
    # Vẽ biểu đồ và tính thống kê cho từng nhóm
    for group_col, group_name in groups.items():
        # Vẽ biểu đồ
        plt.figure()
        plot_metrics_by_group(df, group_col, metrics,
                            f'Các chỉ số trung bình theo {group_name}')
        
        # Tính thống kê
        stats = df.groupby(group_col)[metrics].agg(['mean', 'std', 'count'])
        summary_stats[group_col] = stats.to_dict()
    
    return summary_stats

def print_detailed_statistics(data, group_col, metrics):
    """
    In thống kê chi tiết cho một nhóm cụ thể
    """
    stats = data.groupby(group_col)[metrics].agg([
        'mean', 'std', 'min', 'max', 'count'
    ]).round(2)
    
    return stats
```

#### 5.1 Feature Encoding

```python
# Feature Encoding bằng One-Hot Encoding 
from sklearn.preprocessing import OneHotEncoder

# Các cột phân loại cần mã hóa
cate_col = ['CampaignChannel', 'CampaignType']

# Khởi tạo One-Hot Encoding 
encod = OneHotEncoder(sparse_output=False)  #trả về ma trận thưa 

# Mã hóa
encoded_data = encod.fit_transform(df[cate_col])

# Lấy tên các đặc trưng sau khi mã hóa
feature_names = encod.get_feature_names_out(cate_col)

# Lưu mã hóa trong DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=feature_names)

# Ghép kết quả mã hóa thành 1 chuỗi
df['CampaignType_encoded'] = encoded_df[[col for col in encoded_df.columns if 'CampaignType' in col]].astype(int).astype(str).agg(''.join, axis=1).astype(int)
df['CampaignChannel_encoded'] = encoded_df[[col for col in encoded_df.columns if 'CampaignChannel' in col]].astype(int).astype(str).agg(''.join, axis=1).astype(int)
# tạo từ điển lưu trữ kết quả mã hóa
mapping_dict={}

# Lưu ánh xạ cho mỗi cột 
for col in cate_col:
    # Lấy giá trị phân loại ban đầu 
    values=df[col].unique()
    # lấy chuỗi mã hóa tương ứng
    encoded_values=df[col + '_encoded'].unique()
    # ánh xạ 
    mapping_dict[col]={original: encoded for original, encoded in zip(values, encoded_values)}
    

# Xóa các cột phân loại ban đầu 
df=df.drop(columns = cate_col )

print(mapping_dict)
```
